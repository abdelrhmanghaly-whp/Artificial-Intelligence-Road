```python
import numpy as np
# best case at the lowest cost approaching 0
# htfdl t8er fel learning rate, number of it, l7ad ma twsl ll min cost
def gradientDescent(x,y):
    mCurr = bCurr = 0
    it = 1000
    numberofData = len(x)
    learningRate = 0.01
    for i in range(it):
        y_predicte = mCurr * x + bCurr
        realValue=y-y_predicte
        cost = (1/numberofData) * sum([val**2 for val in (realValue)])
        mDerivative = -(2/numberofData)*sum(x*(realValue))
        bDerivative = -(2/numberofData)*sum(realValue)
        mCurr -= learningRate * mDerivative
        bCurr -= learningRate * bDerivative
        print(" m = {} \n b = {} \n cost = {} \n it = {} \n -----------------------------".format(mCurr,bCurr,cost,i))

x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
gradientDescent(x,y)
```
