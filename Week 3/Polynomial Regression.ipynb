```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import os

os.chdir(r"C:\Users\N..C..C\Downloads\Documents")
```

```python
data = pd.read_csv("positionSalaries.csv")
x = data.iloc[:, 1:2].values
y = data.iloc[:, 2].values
print(data)
```

```python
Regression = LinearRegression()
Regression.fit(x,y)
y_pred = Regression.predict(x)
plt.scatter(x, y, color="blue")
plt.plot(x, y_pred, color="green")
plt.title("Linear Regression")
plt.xlabel("Position Level")
plt.ylabel("Salary")
```

```python
pRegression = PolynomialFeatures(degree = 4)
xPol = pRegression.fit_transform(x)
afterPRegression = LinearRegression()
afterPRegression.fit(xPol, y)
xPol_pred = afterPRegression.predict(xPol)
plt.scatter(x, y, color="blue")
plt.plot(x, xPol_pred, color="green")
plt.title("Linear Regression")
plt.xlabel("Position Level")
plt.ylabel("Salary")
```

```python
x_grid = np.arange(min(x), max(x), 0.1)
print(x_grid)
print("=========================================================================")
x_grid = x_grid.reshape((len(x_grid)),1)
x_grid
```

```python
plt.scatter(x, y, color="blue")
plt.plot(x_grid, afterPRegression.predict(pRegression.fit_transform(x_grid)), color="green")
plt.title("Linear Regression")
plt.xlabel("Position Level")
plt.ylabel("Salary")
```
